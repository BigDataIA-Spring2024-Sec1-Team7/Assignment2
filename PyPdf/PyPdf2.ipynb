```python
import os
from PyPDF2 import PdfReader

folder_path = "/content/drive/MyDrive/abc"
output_folder = "/content/drive/MyDrive/pyPDF"

def get_pdf_files(folder_path):

  pdf_files = []
  for filename in os.listdir(folder_path):
    if filename.endswith(".pdf"):
      pdf_files.append(os.path.join(folder_path, filename))
  return pdf_files



pdf_files = get_pdf_files(folder_path)

print(f"Found {len(pdf_files)} PDF files:")

for filename in range(0, len(pdf_files)):
  reader = PdfReader(pdf_files[filename])

  fname = os.path.basename(pdf_files[filename])
  year = fname[0:4]
  level = fname[5:7]
  print(level)
  print(pdf_files[filename])

  output_filename = f"PyPDF_RR_{year}_{level}_combined.txt"
  output_filepath = os.path.join(output_folder, output_filename)

  with open(output_filepath, "w") as f:
    for i in range(0, len(reader.pages)):
      page = reader.pages[i]
      f.write(page.extract_text())


  print(f"Extracted text from '{filename}' and saved to: {output_filepath}")
```
